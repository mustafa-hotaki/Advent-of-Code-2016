A little messy but gets the job done. I may refine the data structure a bit.


```
data = []
with open('inputs_day_10.txt', 'r') as f:
  for line in f:
    data.append(line.strip())

data[:10]
```

```
# The input data gives us the initial values and mappings of bots to bots or bots to outputs
# bots_and_outputs -> dic of bots or outputs (keys : 'bot 1', 'output 2', ...)
# bot : {values : [], output: (low, high))

bots_and_outputs = {}

# Create data structure and initialize mappings
for line in data:
  if line.startswith('bot'):
    bot_id = line[ : line.find('gives') - 1]
    low_id = ' '.join(line.split(' ')[5 : 7])
    high_id = ' '.join(line.split(' ')[10 : 12])
    #print(line, '\n', bot_id, low_id, high_id)
    bots_and_outputs[bot_id] = {'values' : [], 'outputs' : (low_id, high_id)}

    if not low_id in bots_and_outputs: bots_and_outputs[low_id] = {'values' : []}
    if not high_id in bots_and_outputs: bots_and_outputs[high_id] = {'values' : []}

# Initial values
for line in data:
  if line.startswith('value'):
    bot_id = line[line.find('to') + 3 : ]
    value = int(line[line.find(' ') + 1 : line.find('goes') - 1])
    #print(line, '\n', bot_id, value)
    bots_and_outputs[bot_id]['values'].append(value)

bots_and_outputs
```

# Part 1

```
while(True):
  has_two = []
  for k in bots_and_outputs:
    if k.startswith('bot'):
      if len(bots_and_outputs[k]['values']) == 2:
        #print('Found', k)
        has_two.append(k)

  if len(has_two) == 0: # Only 1 bot has two items?
    break

  #print(has_two)
  for found in has_two:

    #found = has_two[0]

    if 61 in bots_and_outputs[found]['values'] and 17 in bots_and_outputs[found]['values']:
      print('Responsible', found)

    low_v = min(bots_and_outputs[found]['values'])
    high_v = max(bots_and_outputs[found]['values'])

    # Clear
    bots_and_outputs[found]['values'] = []

    bots_and_outputs[bots_and_outputs[found]['outputs'][0]]['values'].append(low_v)
    bots_and_outputs[bots_and_outputs[found]['outputs'][1]]['values'].append(high_v)

bots_and_outputs
```
