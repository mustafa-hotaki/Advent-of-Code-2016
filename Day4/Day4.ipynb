# Advent of Code 2016 - Day 4

```
data = []
with open('inputs_day_4.txt', 'r') as f:
  for line in f:
    data.append(line.strip())

data[:5]
```

```
# Parse
rooms = []
for x in data:
  a = x.split('[')[0]
  encrypted_name = x[ : x.rfind('-')]
  sector_id = int(x[x.rfind('-') + 1: x.find('[')])
  checksum = x[x.find('[') + 1: x.find(']')]

  rooms.append((encrypted_name, sector_id, checksum))

rooms[:5]
```

# Part 1

```
from collections import Counter

valid_sector_ids_sum = 0
valid_rooms = []
for encrypted_name, sector_id, checksum in rooms:

  # I could probably do a more elegant solution, but I can't think right now, so here is an ugly one
  #print(encrypted_name, checksum)
  encrypted_name_string = ''.join(encrypted_name.split('-')) # Removes dashes
  c = Counter(encrypted_name_string)
  a = c.most_common() # Looks like: [('d', 4), ('h', 3), ('k', 3) ...]

  reverse = {}
  for i in range(a[0][1], 0, -1):
    for j in a:
      if j[1] == i:
        if i in reverse:
          reverse[i] += [j[0]]
        else:
          reverse[i] = [j[0]]

  target_permutation = ''
  for k in sorted(reverse, reverse=True):
    #print(k)
    target_permutation += ''.join(sorted(reverse[k]))

  target_permutation_top_five = target_permutation[:5]

  if target_permutation_top_five == checksum:
    valid_rooms.append((encrypted_name, sector_id, checksum))
    valid_sector_ids_sum += sector_id

  #print(target_permutation_top_five)

valid_sector_ids_sum
```

## Part 2

Learned about [Ceasar Cipher](#https://en.wikipedia.org/wiki/Caesar_cipher)

```
from string import ascii_lowercase

def decrypt(string, shift):
  return ''.join([' ' if c == '-' else ascii_lowercase[(ascii_lowercase.find(c) + shift) % 26] for c in string])

a = 'qzmt-zixmtkozy-ivhz'
decrypt(a, 343)
```

```
for encrypted_name, sector_id, checksum in valid_rooms:
  decrypted = decrypt(encrypted_name, sector_id)
  print(sector_id, encrypted_name, decrypted)

# 482 zadftbaxq-anvqof-efadmsq northpole object storage
```
